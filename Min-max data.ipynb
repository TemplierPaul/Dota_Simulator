```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
import json
with open('features_list.json', 'r') as fp:
    features_list = json.load(fp)
features_list
```

```python
pd.Series(features_list)
```

```python
path = "games_data/*"
files = glob.glob(path)
print(len(files), 'Files to load')
```

```python
df = pd.DataFrame()

for f in files:
    df_file = pd.read_csv(f, index_col=0)
    df=pd.concat([df, df_file.drop(columns="action")], axis=0)
    print('%s   \tImported: %d lines' % (f, len(df)), end='\n')
    
df.shape
```

```python
des = df.describe()
des
```

```python
mm = pd.DataFrame()
mm["min"]=des.loc['min']
mm["max"]=des.loc['max']
mm['Name']=pd.Series(features_list)
```

```python
mm.to_csv("min_max.csv")
```
